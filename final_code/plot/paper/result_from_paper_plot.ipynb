```python
import openpyxl
import numpy as np
import matplotlib.pyplot as plt
```

```python
wb1 = openpyxl.load_workbook('result_from_paper.xlsx')
```

```python
data = np.zeros((36, 5))
for i in range(3, 39):
    for j in range(4, 7):
        data[i-3][j-4] = wb1["Sheet2"].cell(row = i, column = j).value
    data[i-3][3] = 2 * data[i-3][1] * data[i-3][2] / (data[i-3][1] + data[i-3][2])
    data[i-3][4] = np.mean(np.array([data[i-3][0], data[i-3][1], data[i-3][2], data[i-3][3]]))
```

```python
algorithm_index = ["Paper(Zimmermann2011)", "LogisticRegression", "RF with SMOTEENN", "RF with SMOTE"]
```

```python
x_index = ["2.0", "2.1", "3.0"]
```

```python
y_index = ["Accuracy Score", "Recall Score", "Precision Score", "F1 Score", "Mean of the above four Scores"]
```

```python
plt.figure(figsize = (100, 50))
for i in range(3):
    for j in range(5):
        plt.subplot(3, 5, 5 * i + j + 1)
        for k in range(4):
            plt.plot(x_index, data[[12*i+k, 12*i+k+4, 12*i+k+8], j], ".-")
        plt.legend(algorithm_index, bbox_to_anchor=(1,1))
        plt.xlabel("Test Set")
        plt.ylabel(y_index[j])
        plt.title(y_index[j] + " (" + x_index[i] + " as train set)")
plt.savefig('all.png', dpi=150)    
```
