```python
import openpyxl
import numpy as np
import matplotlib.pyplot as plt
```

```python
wb1 = openpyxl.load_workbook('result_from_website.xlsx')
```

```python
algorithm_index = ["LogisticRegression", "MLP with OverSample", "RF with OverSample", "RF with SMOTEENN", "RF with SMOTE", "RF"]
```

```python
data = np.zeros((6, 25, 5))
for i in range(6):
    for j in range(2,27):
        for k in range(3,8):
            data[i][j-2][k-3] = wb1[algorithm_index[i]].cell(row = j, column = k).value
```

```python
x_index = ["Eclipse_JDT_Core", "Eclipse_PDE_UI", "Equinox_Framework", "Lucene", "Mylyn"]
```

```python
y_index = ["Accuracy Score", "Recall Score", "Precision Score", "F1 Score", "Mean of the above four Scores"]
```

```python
plt.figure(figsize = (100, 50))
for i in range(5):
    for j in range(5):
        plt.subplot(5, 5, 5 * i + j + 1)
        for k in range(6):
            plt.plot(x_index, data[k, 5*i:5*i+5, j], ".-")
        plt.legend(algorithm_index, bbox_to_anchor=(1,1))
        plt.xlabel("Test Set")
        plt.ylabel(y_index[j])
        plt.title(y_index[j] + " (" + x_index[i] + " as train set)")
plt.savefig('all.png', dpi=150)
```
