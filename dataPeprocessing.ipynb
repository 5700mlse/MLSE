```python
import pandas as pd

def split_the_csv(matrix_path, filename):

    def load_one_version_method_matrix(matrix_version):
        csv_path = matrix_path+'/'+ matrix_version
        return pd.read_csv(csv_path)
    change_metrics = load_one_version_method_matrix(filename)
    print(change_metrics.info())

    type(change_metrics)
    columnlist = change_metrics.columns.values[0].replace('\'','')
    columnlist  = columnlist.replace(': ','')
    columnlist = columnlist.replace(' ','')
    columnlist = columnlist.split(";")
#     print(columnlist)

    Name = change_metrics.columns.values[0]
#     print(Name)

#     print(change_metrics[Name])

    change_metrics= pd.DataFrame(change_metrics[Name].str.split(';',0).tolist(),columns = columnlist)
    change_metrics.to_csv(matrix_path+"/modified_"+filename)
```

```python
matrix_path1 = 'AlltheMatrix/differentSoftware/eclipse/'
matrix_path2 = 'AlltheMatrix/differentSoftware/equinox/'
matrix_path3 = 'AlltheMatrix/differentSoftware/lucene/'
matrix_path4 = 'AlltheMatrix/differentSoftware/mylyn/'
matrix_path5 = 'AlltheMatrix/differentSoftware/pde/'
matrix_path_list = [matrix_path1,matrix_path2,matrix_path3,matrix_path4,matrix_path5]
filename = "single-version-ck-oo.csv"
for i in matrix_path_list:
    split_the_csv(i,filename)
```

```python
import xlrd
import pandas as pd
from pandas import DataFrame
for i in matrix_path_list:
    data1 = pd.read_csv(i+'modified_bug-metrics.csv')
    data2 = pd.read_csv(i+'modified_change-metrics.csv')
    data3 = pd.read_csv(i+'modified_complexity-code-change.csv')
    data4 = pd.read_csv(i+'modified_single-version-ck-oo.csv')
    df1 = pd.DataFrame(data1)
    df2 = pd.DataFrame(data2)
    df3 = pd.DataFrame(data3)
    df4 = pd.DataFrame(data4)
    writer = pd.ExcelWriter('data.xlsx')
    df1.to_excel(writer, sheet_name = 'bug-metrics')
    df2.to_excel(writer, sheet_name = 'change-metrics')
    df3.to_excel(writer, sheet_name = 'complexity-code-change')
    df4.to_excel(writer, sheet_name = 'single-version-ck-oo')
    writer.save()
    writer.close()
    wb = xlrd.open_workbook('data.xlsx') 
    d1 = pd.read_excel('data.xlsx', sheet_name = 'bug-metrics', usecols = [i for i in range(2, 8)], encoding='utf8')
    d2 = pd.read_excel('data.xlsx', sheet_name = 'change-metrics', usecols = [i for i in range(3, 18)], encoding='utf8')
    d3 = pd.read_excel('data.xlsx', sheet_name = 'complexity-code-change', usecols = [i for i in range(3, 8)], encoding='utf8')
    d4 = pd.read_excel('data.xlsx', sheet_name = 'single-version-ck-oo', usecols = [i for i in range(3, 26)], encoding='utf8')
    all_data = pd.concat([d1, d2, d3, d4], axis = 1)
    pd.DataFrame(all_data).to_csv(i+'all_data.csv')
```
