```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.random.rand(1000) * np.pi
b = -1 + np.random.rand(1000) * 2
```

```python
m = 0
n = 0
for i in range(1000):
    if b[i] <= np.sin(a[i] - np.pi/2):
        if m == 0:
            plt.plot(a[i]/np.pi, (b[i] + 1)/2, 'r.', label = 'False Samples')
            m += 1
        plt.plot(a[i]/np.pi, (b[i] + 1)/2, 'r.')
    else:
        if n == 0:
            plt.plot(a[i]/np.pi, (b[i] + 1)/2, 'b.', label = 'True Samples')
            n += 1
        plt.plot(a[i]/np.pi, (b[i] + 1)/2, 'b.')
x = np.linspace(0, np.pi, 100)
plt.plot(x/np.pi, (np.sin(x-np.pi/2) + 1)/2, 'g', label = 'Decision Boundary', linewidth = 3)
plt.title("Logistic Regression Example")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(bbox_to_anchor = (1, 1))
plt.savefig("lr", dpi=200, bbox_inches='tight')
```
